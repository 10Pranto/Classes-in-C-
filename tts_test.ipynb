```python
import os
import soundfile as sf
import pandas as pd
from datasets import load_dataset
from tqdm import tqdm

dataset = load_dataset("google/fleurs", "bn_in", split="train", streaming=True)

output_base_dir = "fleurs_bn_data"
output_wav_dir = os.path.join(output_base_dir, "wavs")
os.makedirs(output_wav_dir, exist_ok=True)
filenames = []
texts = []

for data in tqdm(dataset):
  # print(data)
  filename = f"bn_{data['id']}.wav"
  audio = data['audio']['array']
  output_wav_name = os.path.join(output_wav_dir, filename)
  sf.write(output_wav_name, audio, 16000)
  filenames.append(filename.replace('.wav', ''))
  transcription = data['transcription']
  texts.append(transcription)

df = pd.DataFrame()
df['filename'] = filenames
df['text'] = texts

metadata_output_path = os.path.join(output_base_dir, 'metadata.csv')
df.to_csv(metadata_output_path, header=False, index=False)

```
